```python
# Grafico de precios por metro cuadrado para los barrios de Bogota (Correcto)

# Creamos una nueva columna en la tabla
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area'] 
inmuebles.head(3)
#inmuebles.info()
inmuebles.groupby('Barrio').mean()  # Agrupamos por barrio y sacamos el promedio de las columnas numericas para cada barrio

# Vamos a sumar todos los precios de los inmuebles de cada barrio y dividirlos entre la sumatoria de todas las areas de los inmuebles de cada barrio
datos_barrios = inmuebles.groupby('Barrio').sum()
datos_barrios

datos_barrios['Valor_m2_Barrio'] = datos_barrios['Precio_Millon']/datos_barrios['Area']
datos_barrios

# Sustituiremos los valores correctos obtenidos en valor_m2_barrio en el dataframe inmuebles original
m2_barrio = dict(datos_barrios['Valor_m2_Barrio'])

inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
#Se creo un diccionario con los valores de valor_m2_barrio y para cada valor de la columna barrio se sustituyo por el valor del diccionario
inmuebles.head(3)

# Ahora tomaremos los 10 barrios con mas inmuebles
top_barrios = inmuebles['Barrio'].value_counts()[:10].index   # Nos muestra los 10 barrios con mas inmuebles
datos_barrios.reset_index(inplace=True) # Con esto a datos_barrios le agregamos una columna con los indices de los barrios y ya no solo como indice
datos_barrios

datos_barrios.query('Barrio in @top_barrios')   # Nos muestra los datos de los 10 barrios con mas inmuebles

# Creamos nuestro lienzo
plt.figure(figsize=(10,8))      # figsize es para el tamaño del lienzo / 10 es el ancho y 8 es el alto
ax = sns.barplot(x="Barrio", y="Valor_m2_Barrio", data = datos_barrios.query('Barrio in @top_barrios'))

# Personalizacion del grafico
ax.set_xlabel('Barrios de bogota', fontsize=13, color='Teal')    # Etiqueta del eje x
ax.set_ylabel('Precio por m^2 (Millones $)', fontsize = 13, color='Teal')    # Etiqueta del eje y
ax.set_title('Top10 Barrios Bogota (Valor m^2 en Millones)', fontsize=16,
                        fontweight='bold', color='Teal', loc='center')    # Titulo del grafico
ax.tick_params(axis='x', rotation=90)    # Rotamos las etiquetas del eje x 90 grados

plt.show()      # Mostramos el grafico
```

```python
plt.figure(figsize=(10,8))      # figsize es para el tamaño del lienzo / 10 es el ancho y 8 es el alto
ax2 = sns.boxplot(x="Barrio", y="Valor_m2_Millon", data = inmuebles.query('Barrio in @top_barrios'))
# El boxplot nos muestra la distribucion de los datos / en X y Y se pasan las columnas a graficar, no son titulos

# Personalizacion del grafico
ax2.set_xlabel('Barrios de bogota', fontsize=13, color='Teal')    # Etiqueta del eje x
ax2.set_ylabel('Precio por m^2 (Millones $)', fontsize = 13, color='Teal')    # Etiqueta del eje y
ax2.set_title('Boxplot Barrios Bogota (Valor m^2 en Millones)', fontsize=16,
                        fontweight='bold', color='Teal', loc='center')    # Titulo del grafico
ax2.tick_params(axis='x', rotation=45)    # Rotamos las etiquetas del eje x 90 grados

plt.show()      # Mostramos el grafico

# Las cajas nos muestran el rango intercuartilico (25%(1ra linea) - 75%(ultima linea))
# Los puntos fuera de la caja son los outliers(los datos que estan muy alejados de la mayoria de los datos) / siempre hay que tenerlos en cuenta
# La linea central es la mediana
# La linea horizontal es la media
```

```python
# Agruparemos por estrato(Nombre UPZ: conjunto barrios de zona geografica) algunas de las caracteristicas mas relevantes
datos_dane.groupby('NOMBRE_ESTRATO')[['CONJUNTO_CERRADO','INSEGURIDAD','TERMINALES_BUS','BARES_DISCO','RUIDO','OSCURO_PELIGROSO','SALARIO_MES','TIENE_ESCRITURA','PERDIDA_TRABAJO_C19','PERDIDA_INGRESOS_C19','PLANES_ADQUIRIR_VIVIENDA']].mean().head()

# Ya que las columnas nos arrojaban preguntas de si o no(1=si, 2=no), entonces todos los 2 los cambiaremos a 0 y asi evitar que en los promedios me den valores decimales
datos_aux = datos_dane[['NOMBRE_ESTRATO','CONJUNTO_CERRADO','INSEGURIDAD','TERMINALES_BUS','BARES_DISCO','RUIDO','OSCURO_PELIGROSO','SALARIO_MES','TIENE_ESCRITURA','PERDIDA_TRABAJO_C19','PERDIDA_INGRESOS_C19','PLANES_ADQUIRIR_VIVIENDA']].replace(2,0)
datos_aux

# Revisamos si no se cambio algun nombre de estrato que incluyera un 2 y le pusiera ahora un 0
datos_aux.loc[datos_aux.NOMBRE_ESTRATO == '20 de Julio']
# Comprobamos que no se cambio ningun nombre porque el replace fue numerico y no de string como "2","0"

# Agrupamos por estrato y sacamos el promedio de las columnas que nos interesan ahora si con porcentajes correctos de cuantos dijeron si
datos_tratados = datos_aux.groupby('NOMBRE_ESTRATO')[['CONJUNTO_CERRADO','INSEGURIDAD','TERMINALES_BUS','BARES_DISCO','RUIDO','OSCURO_PELIGROSO','SALARIO_MES','TIENE_ESCRITURA','PERDIDA_TRABAJO_C19','PERDIDA_INGRESOS_C19','PLANES_ADQUIRIR_VIVIENDA']].mean()
datos_tratados
```

```python
# Ahora vamos a fusionar(merge) los datos de inmuebles con los de la encuesta de DANE
pd.merge(inmuebles,datos_tratados, left_on='UPZ', right_on='NOMBRE_ESTRATO', how='left')
# fusionamos por la columna UPZ(left on) de inmuebles con la columna NOMBRE_ESTRATO(right on) de datos_tratados
# how left: nos quedamos con todos los datos de la tabla izquierda, es decir, de inmuebles

# Guardamos este nuevo dataframe en una variable
datos_ml = pd.merge(inmuebles,datos_tratados, left_on='UPZ', right_on='NOMBRE_ESTRATO', how='left') # ML: Machine Learning
datos_ml.info()
```

```python
''' Para el modelo que vamos a crear no podemos tener elementos de tipo objeto como lo es UPZ, entonces ya que
    UPZ tiene un codigo asignado, vamos a traerlo del archivo que ya tenemos y poder fusionarlo con los datos '''

# Cargamos el archivo de UPZ
upz = pd.read_csv('/content/drive/MyDrive/Python_data_science_curso/cod_upz.csv')
upz.head()
datos_ml = pd.merge(datos_ml,upz,left_on='UPZ',right_on='NOMBRE_ESTRATO', how='inner')
# how inner: nos quedamos solo con los datos que esten en ambas tablas en caso de que haya datos que no esten en ambas tablas
```

```python
""" Por precaucion colocamos estos comandos en una nueva celda ya que estar ejecutando varias veces el codigo
    anterior, hace merge sobre merge y nos incrementa las columnas """
datos_ml.head()
#datos_ml.shape  # (9318, 26)
datos_ml.info()
# Nos damos cuenta que de los 9318 datos que tenemos, todas las columnas contienen 9318 datos no nulos, por lo tanto tenemos un dataframe limpio para la ML
```

```python
#Ahora vamos a crear un modelo de regresion lineal para predecir el precio de los inmuebles
from sklearn.linear_model import LinearRegression   # Importamos el modelo de regresion lineal
from sklearn.model_selection import train_test_split    # Importamos la funcion para dividir los datos en entrenamiento y datos para prueba

# Asignamos datos de entrada (X) y datos de salida (y)(Que es lo que queremos predecir) / Base Line
X = datos_ml[['COD_UPZ_GRUPO']] # X mayus porque por lo general la entrada es una matriz
y = datos_ml['Precio_Millon']

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=99)
# train_test_split: recibe datos entrada X y datos salida y
# test_size: porcentaje de datos que se van a usar para prueba por lo que el resto (75%) es para entrenamiento / por defecto es 0.25
# random_state: semilla para que los datos de prueba sean los mismos cada vez que se ejecute el codigo con ese estado / se puede poner cualquier numero
```

```python
# Instanciamos el modelo de regresion lineal
modelo = LinearRegression()
```

```python
# Realizamos el entrenamiento del modelo
modelo.fit(X_train, y_train)    # Aqui ya se aplicó machine learning
```

```python
# Realizamos la prediccion
y_predict_test = modelo.predict(X_test)  # A partir de los datos prueba X_test, se predice el precio de los inmuebles y_predict_test
```

```python
# Validacion del modelo
from sklearn.metrics import mean_absolute_error, r2_score
# mean_absolute_error: Toma diferencias de valores absolutos entre los valores reales y los valores predichos
# r2_score: Coeficiente de determinacion R2 para observar que tan bueno es el modelo / 0 es el peor modelo y 1 es el mejor modelo

baseline_mae = mean_absolute_error(y_test, y_predict_test)  # Error absoluto medio entre los valores reales y los valores predichos
baseline_r2 = r2_score(y_test, y_predict_test)  # Coeficiente de determinacion R2 entre los valores reales y los valores predichos
print(baseline_mae,baseline_r2)

# Observamos que el modelo no es bueno debido a que solo tomamos un atributo que fue COD_UPZ_GRUPO, mientras mas tomemos mejor sera el modelo
```

```python
# Vamos a mejorar nuestro modelo agregando mas atributos a la entrada X

X = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']] 
y = datos_ml["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```
